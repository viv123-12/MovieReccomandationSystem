```python
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['overview'] = movies['overview'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
#converts the tags into string
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'].apply(lambda x:" ".join(x))
```

```python
from nltk import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    L = []
    for i in text.split():
        L.append(ps.stem(i))
    return " ".join(L);
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 5000 , stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
print(vectors.shape)
```

```python
cv.get_feature_names()
```

```python
# find cosine similarity which is an angle between two vectors whereas in euclidian distance in tip to tip diastance
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
# print(similarity.shape)
sorted(list(enumerate(similarity[0])),reverse = True,key=lambda x:x[1])[0:6]
```

```python
def recommand(movie):
    index = new_df[(new_df['title'] == movie)].index[0] 
    movie_list = sorted(list(enumerate(similarity[index])),reverse = True,key=lambda x:x[1])[0:6] 
    return movie_list
```
